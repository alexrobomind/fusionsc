```python
import fsc
from fsc.devices import w7x
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pyvista as pv
```

```python
fsc.resolve.importOfflineData("../w7x.fsc")
```

```python
root = fsc.native.connectSameThread()
```

```python
camProvider = root.newHFCamProvider().service
```

```python
np.radians(20)
```

```python
proj = camProvider.makeToroidalProjection(800, 600, 0, 5.5, 0, 0.3, 5, 2, 0.3).wait()
```

```python
proj
```

```python
cam = camProvider.makeCamera(proj, w7x.divertor().resolve().wait().geometry, 0.5, 0.001).wait()
```

```python
plt.figure(figsize = (12, 9))
plt.imshow(np.asarray(cam.determinantBuffer).T[::-1])
plt.colorbar()
```

```python
plt.figure(figsize = (12, 9))
plt.imshow(np.asarray(cam.depthBuffer).T[::-1])
```
